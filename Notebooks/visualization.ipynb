# Visualization

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import folium
import webbrowser
```

```python
data = pd.read_csv("Data/us_daily_processed.csv")
data.head()
```

```python
d_test=data.groupby(["Province_State","date"])[['Confirmed','Deaths']].sum().reset_index()
d_test.head()
```

```python
d_total = d_test.groupby(["Province_State"])[['Confirmed','Deaths']].sum().reset_index()
nd = np.shape(d_total)[0]
d_total
```

```python
cluster_data = pd.read_csv("Data/useful_features_by_state.csv")
```

```python
cluster_data.columns
```

```python
cluster_data
```

```python
states = list(set(cluster_data['state']))
```

```python
for i in range(nd):
    item = d_total.loc[i,'Province_State']
    if item not in states:
        d_total = d_total.drop([i],axis=0)
d_total = d_total.reset_index().drop(['index'],axis=1)
```

```python
d_total
```

```python
temp = cluster_data
temp['2019'] = temp['2019'].astype(float)
temp = pd.merge(temp, d_total, how='outer', left_on='state',right_on='Province_State')
```

```python
temp['Death_rate'] = temp['Deaths']/temp['Confirmed']
```

```python
state_geo = 'Data/us-states.json'
```

```python
temp
```

```python
#Population and confirmed case
m = folium.Map(location=[0, 0], tiles='cartodbpositron',
               min_zoom=1, max_zoom=8, zoom_start=1.5)

m.choropleth(geo_data=state_geo, name='choropleth',data=temp,
             columns=['State','2019'],
             key_on='feature.id',
             # 'BuGn', 'BuPu', 'GnBu', 'OrRd', 'PuBu', 'PuBuGn', 'PuRd', 'RdPu','YlGn', 'YlGnBu', 'YlOrBr', and 'YlOrRd'.
             fill_color='OrRd', fill_opacity=0.7, line_opacity=0.1,
             legend_name="Population")

m

for i in range(0, len(temp)):
    folium.Circle(
        location=[temp.iloc[i]['Latitude'], temp.iloc[i]['Longitude']],
        color='crimson', fill='crimson',
        tooltip =   '<li><bold>State : '+str(temp.iloc[i]['state'])+
                    '<li><bold>Polutation : '+str(temp.iloc[i]['2019'])+
                    '<li><bold>Total confirmed : '+str(temp.iloc[i]['Confirmed'])
        ,
        radius=int(temp.iloc[i]['Confirmed']*0.01)).add_to(m)

#Save 
m.save('Plots/Population_and_confirmed_case.html') 
#Show
webbrowser.open("Plots/Population_and_confirmed_case.html")
```

The first map we got shows the Population (published by census in 2019) and confirmed cases of each state. In our map, The different color corresponds to different levels of populations and the radius of the circle is proportional to the number of confirmed case. And when we move the mouse to the circle, we can see the exact value.<p> Generally speaking, states with smaller populations also have fewer confirmed case. And the situation in New York is the worst.

```python
#Condirmed case and death case
m = folium.Map(location=[0, 0], tiles='cartodbpositron',
               min_zoom=1, max_zoom=8, zoom_start=1.5)

m.choropleth(geo_data=state_geo, name='choropleth',data=temp,
             columns=['State','Confirmed'],
             key_on='feature.id',
             # 'BuGn', 'BuPu', 'GnBu', 'OrRd', 'PuBu', 'PuBuGn', 'PuRd', 'RdPu','YlGn', 'YlGnBu', 'YlOrBr', and 'YlOrRd'.
             fill_color='RdPu', fill_opacity=0.7, line_opacity=0.1,
             legend_name="Confirmed cases",
             threshold_scale =[18192,100000,1000000,5000000,10000000,14860849])


m

for i in range(0, len(temp)):
    folium.Circle(
        location=[temp.iloc[i]['Latitude'], temp.iloc[i]['Longitude']],
        color='crimson', fill='crimson',
        tooltip =   '<li><bold>State : '+str(temp.iloc[i]['state'])+
                    '<li><bold>Total confirmed : '+str(temp.iloc[i]['Confirmed'])+
                     '<li><bold>Total deaths : '+str(temp.iloc[i]['Deaths'])
        ,
        radius=int(temp.iloc[i]['Deaths']*0.2)).add_to(m)

#Save 
m.save('Plots/Condirmed_case_and_death_case.html') 
#Show
webbrowser.open("Plots/Condirmed_case_and_death_case.html")
```

This map shows different levels of confirmed cases and death cases for each state. We can see that New York and New Jersey has the most confirmed cases while Massachusetts, Connecticut, Pennsylvania, Maryland, Michigan, Illinois, Texas, Louisiana, Florida, Georgia and California also has more than 1000000 confirmed cases each.<p>
Generally speaking, the epidemic is more serious in coastal states.

```python
#Hopsital Beds and Death Rate
m = folium.Map(location=[0, 0], tiles='cartodbpositron',
               min_zoom=1, max_zoom=8, zoom_start=1.5)

m.choropleth(geo_data=state_geo, name='choropleth',data=temp,
             columns=['State','beds'],
             key_on='feature.id',
             # 'BuGn', 'BuPu', 'GnBu', 'OrRd', 'PuBu', 'PuBuGn', 'PuRd', 'RdPu','YlGn', 'YlGnBu', 'YlOrBr', and 'YlOrRd'.
             fill_color='GnBu', fill_opacity=0.7, line_opacity=0.1,
             legend_name="Hospital beds (per 1000 residents)")

m

for i in range(0, len(temp)):
    folium.Circle(
        location=[temp.iloc[i]['Latitude'], temp.iloc[i]['Longitude']],
        color='crimson', fill='crimson',
        tooltip =   '<li><bold>State : '+str(temp.iloc[i]['state'])+
                    '<li><bold>Hospital beds (per 1000 residents) : '+str(temp.iloc[i]['beds'])+
                    '<li><bold>Death_rate : '+str(temp.iloc[i]['Death_rate'])
        ,
        radius=int(temp.iloc[i]['Death_rate']*500000)).add_to(m)

#Save 
m.save('Plots/Hopsital_Beds_and_Death_Rate.html') 
#Show
webbrowser.open("Plots/Hopsital_Beds_and_Death_Rate.html")
```

We try to use hospital beds holds for per 1000 residents in each state to represent corresponding local medical resourse. <p>
For state with high hospital beds level like Texas and Virginia, although it has much confirmed case, but its death rate is not so high.
